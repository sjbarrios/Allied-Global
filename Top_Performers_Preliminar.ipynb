### Empleados.csv

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
empleados = pd.read_csv('D:/OneDrive - Universidad Galileo/Data Science/PAPD/Data Science at Work/Datos/Empleados.csv')
empleados.drop(['id_unidad','Puesto.1'], axis=1, inplace=True)
empleados
```

### empleados_vacaciones.csv

```python
vacaciones = pd.read_csv('D:/OneDrive - Universidad Galileo/Data Science/PAPD/Data Science at Work/Datos/empleados_vacaciones.csv')
vacaciones.drop(['departamento','PUESTO','Fecha_Ingreso','PAIS','SITE','ESTATUS'], axis=1, inplace=True)
vacaciones
```

### Top_performers_2017_2020.csv

```python
top_performers = pd.read_csv('D:/OneDrive - Universidad Galileo/Data Science/PAPD/Data Science at Work/Datos/Top_performers_2017_2020.csv')


top_performers.rename(columns={'Year':'Year_TP',
                          'Month':'Month_TP',
                          'Cuenta':'Cuenta_TP'}, 
                 inplace=True)
top_performers
```

### Empleados_amonestaciones.csv

```python
amonestaciones = pd.read_csv('D:/OneDrive - Universidad Galileo/Data Science/PAPD/Data Science at Work/Datos/Empleados_amonestaciones.csv')
amonestaciones.drop(['NombrePais','Cuenta','Estatus_Empleado','PUESTO','SITE'], axis=1, inplace=True)
amonestaciones
```

### Empleados_planificacion_horarios.csv

```python
planificacion_horarios = pd.read_csv('D:/OneDrive - Universidad Galileo/Data Science/PAPD/Data Science at Work/Datos/Empleados_planificacion_horarios.csv')
planificacion_horarios.drop(['Cuenta','pais','estado','site'], axis=1, inplace=True)
planificacion_horarios
```

### Análisis Inicial Top_Performers

```python
inner_join1 = pd.merge(empleados,  
                      top_performers,  
                      on ='Codigo_Empleado',  
                      how ='inner') 
dataframe = pd.merge(inner_join1,  
                      vacaciones,  
                      on ='Codigo_Empleado',  
                      how ='inner')

dataframe
```
